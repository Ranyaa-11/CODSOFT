```python
import tkinter as tk
from tkinter import messagebox

tasks = []

def add_task():
    task = entry.get()
    if task:
        tasks.append(task)
        entry.delete(0, tk.END)
        view_tasks()
    else:
        messagebox.showwarning("Input Error", "Please enter a task.")

def delete_task():
    selected = listbox.curselection()
    if selected:
        task = listbox.get(selected)
        tasks.remove(task)
        view_tasks()
    else:
        messagebox.showwarning("Selection Error", "Please select a task to delete.")

def update_task():
    selected = listbox.curselection()
    new_task = entry.get()
    if selected and new_task:
        index = selected[0]
        tasks[index] = new_task
        entry.delete(0, tk.END)
        view_tasks()
    else:
        messagebox.showwarning("Update Error", "Select a task and enter new text.")

def view_tasks():
    listbox.delete(0, tk.END)
    for task in tasks:
        listbox.insert(tk.END, task)

def exit_app():
    root.destroy()

# Main window
root = tk.Tk()
root.title("Task Manager")
root.geometry("500x550")
root.configure(bg="#f0f0f0")

# Title
title = tk.Label(root, text=" Task Manager", font=("Helvetica", 18, "bold"), bg="#f0f0f0")
title.pack(pady=10)

# Entry Frame
entry_frame = tk.Frame(root, bg="#f0f0f0")
entry_frame.pack(pady=10)

entry = tk.Entry(entry_frame, width=30, font=("Helvetica", 12))
entry.pack(side=tk.LEFT, padx=5)

add_btn = tk.Button(entry_frame, text="Add", command=add_task, bg="#4CAF50", fg="white", font=("Helvetica", 10, "bold"))
add_btn.pack(side=tk.LEFT, padx=5)

update_btn = tk.Button(entry_frame, text="Update", command=update_task, bg="#2196F3", fg="white", font=("Helvetica", 10, "bold"))
update_btn.pack(side=tk.LEFT, padx=5)

# Listbox Frame
list_frame = tk.Frame(root)
list_frame.pack(pady=10)

scrollbar = tk.Scrollbar(list_frame)
scrollbar.pack(side=tk.RIGHT, fill=tk.Y)

listbox = tk.Listbox(list_frame, width=50, height=10, font=("Helvetica", 12), yscrollcommand=scrollbar.set)
listbox.pack(side=tk.LEFT)

scrollbar.config(command=listbox.yview)

# Button Frame
btn_frame = tk.Frame(root, bg="#f0f0f0")
btn_frame.pack(pady=20)

view_btn = tk.Button(btn_frame, text="View Tasks", command=view_tasks, bg="#FF9800", fg="white", font=("Helvetica", 10, "bold"))
view_btn.pack(side=tk.LEFT, padx=10)

delete_btn = tk.Button(btn_frame, text="Delete", command=delete_task, bg="#f44336", fg="white", font=("Helvetica", 10, "bold"))
delete_btn.pack(side=tk.LEFT, padx=10)

exit_btn = tk.Button(btn_frame, text="Exit", command=exit_app, bg="#9E9E9E", fg="white", font=("Helvetica", 10, "bold"))
exit_btn.pack(side=tk.LEFT, padx=10)

# Run the app
root.mainloop()
```
